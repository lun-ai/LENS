```python
import pandas as pd
```

Compare the information gain of partitions by learned strategy to the entropy method

```python
# information gain given by the ILP learned strategy
partition_to_entropy = pd.read_csv('test/partition_to_entropy.csv', index_col=0).fillna(0)

# Entropy calculated from minimal (hypothesis space) reduction ratio
entropy = pd.read_csv('data/entropy.csv', index_col=0)[1:]

display(partition_to_entropy)
display(entropy)
```

Is every partition giving the same information gain as the entropy method?

```python
def compare_rows(row1, row2):
    return row1 == row2

# Apply the comparison row by row
print(partition_to_entropy.apply(lambda row: compare_rows(row, entropy.loc[row.name]), axis=1))
```
