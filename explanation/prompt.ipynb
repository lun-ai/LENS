```python
from explanation.explanation import *
from explanation.load_hf_model import *
from langchain_ollama import OllamaLLM
import json
```

## Define Task Names

The `task_names` list defines the different meta-strategy learning tasks for which explanations will be generated. These tasks likely represent different scenarios or configurations in the meta-strategy learning process.

```python
# Define the task names
task_names = ['isolated_behind', 'partition', 'partition_sizes', 'optimal_partition_sizes']
```

## Generate Explanations using Qwen2.5

This cell configures and runs the explanation generation process using the `krith/qwen2.5-coder-32b-instruct:IQ2_M` model.
- `version`: Specifies the 'instruct' version of the prompt template.
- `model`: Defines the Qwen2.5 model to be used via Ollama.
- `template_path`: Path to the instruction-based prompt template.
- `explanation_path`: Path where the generated explanations will be saved.
- `llm`: Initializes the OllamaLLM with the specified model.
- `instruct_prompt_output`: A function call to generate explanations for the defined `task_names` using the instruction-based prompt, saving 5 samples.

```python
# Define the model and prompt template
version = 'instruct'
model = 'krith/qwen2.5-coder-32b-instruct:IQ2_M'
template_path = f"explanation/explanation_template_{version}.txt"
explanation_path = f"explanation/{model.replace('/', '_')}_explanation_{version}.txt"
llm = OllamaLLM(model=model)

instruct_prompt_output(llm, task_names, template_path, explanation_path, sample_size=5)
```

## Generate Explanations using StarCoder2

This cell configures and runs the explanation generation process using the `hf.co/QuantFactory/starcoder2-15b-instruct-v0.1-GGUF:Q4_K_M` model.
- `version`: Specifies to use the 'instruct' version of the prompt template.
- `model`: Defines the StarCoder2 model to be used via Ollama.
- `llm`: Initializes the OllamaLLM with the specified model.
- `template_path`: Path to the prompt template.
- `explanation_path`: Path where the generated explanations will be saved.
- `user_prompt_output`: A function call to generate explanations for the defined `task_names` using the recommended chat template, saving 5 samples.

```python
model = "hf.co/QuantFactory/starcoder2-15b-instruct-v0.1-GGUF:Q4_K_M"
version = 'instruct'
llm = OllamaLLM(model=model)
template_path = f"explanation/explanation_template_{version}.txt"
explanation_path = f"explanation/{model.replace('/', '_')}_explanation_{version}.txt"

user_prompt_output(llm, task_names, template_path, explanation_path, sample_size=5)
```
